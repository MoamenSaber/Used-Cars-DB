# split data to features and lable

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
x = df.drop('Price' , axis = 1)
x
```

```python
y = df['Price']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x , y , test_size=0.3 , random_state=0)
X_train.shape , X_test.shape ,y_train.shape , y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_Scaled = scaler.transform(X_train)
X_train_Scaled.shape
```

```python
scaler.fit_transform(X_train).shape
```

```python
X_test_Scaled = scaler.transform(X_test)
X_test_Scaled.shape
```

# Modelling

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train_Scaled , np.log(y_train))
```

```python
model.score(X_train_Scaled , np.log(y_train))
```

```python
y_pred = model.predict(X_test_Scaled)
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(np.log(y_test) , y_pred)
```

```python
metrics.mean_absolute_error(np.log(y_test), y_pred)
```

```python
plt.figure(figsize=(30,9))
sns.displot(np.abs(y_test-y_pred))
plt.show()
```
